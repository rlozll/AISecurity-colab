# 수업 실습


```python
## 도미 데이터(도미의 길이와 무게) 준비

bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
import matplotlib.pyplot as plt # matblotlib pyplot 함수를 plt로 줄여서

plt.scatter(bream_length, bream_weight) # 산점도 그래프 그리기
plt.xlabel('length')                    # x축: length
plt.ylabel('weight')                    # y축: weight
plt.show()
```

```python
## 빙어 데이터(빙어의 길이와 무게) 준비
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2,
                12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8,
                12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
## 데이터셋 통합(도미+빙어)
length = bream_length+smelt_length
weight = bream_weight+smelt_weight

print(length)
print(weight)
```

```python
## 점을 찍기 위해 (x, y) 형태로 추출
fish_data = [[l, w] for l, w in zip(length, weight)]
print(fish_data)
```

```python
## 도미(35마리)와 빙어(14마리)로 라벨링
fish_target = [1]*35 + [0]*14
print(fish_target)
```

```python
## k-최근접이웃 알고리즘 사용을 위해 싸이킷런 패키지로부터 임포트
from sklearn.neighbors import KNeighborsClassifier

## k-최근접이웃 분륙 ㅣ선언
kn = KNeighborsClassifier()
```

```python
## fit()은 머신러닝의 훈련 메소드, score()는 평가메소드
# kn 모델에 요소(fish_data)와 레이블(fish_target)을 지정하여 훈련함
kn.fit(fish_data, fish_target)
# kn 모델에 요소(fish_data)와 레이블(fish_target)을 지정하여 평가함
kn.score(fish_data, fish_target)
```

```python
## 산점도 그래프를 다시 그림, 색을 구별하기 위해 두번에 나눠서 출력
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)

## 특정요소가 (30, 600)을 화면에 출력
plt.scatter(30, 600, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
## 특정요소가 (30, 600)의 값을 갖는다면 kn 모델은 무엇으로 예측 하는가?
kn.predict([[30, 600]])
```

```python
## 산점도 그래프 출력: 도미와 빙어 구별하기 위해 원래 데이터로 그림
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)

## 특정요소가 (15, 80)을 화면에 출력
plt.scatter(15, 80, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
## 특정요소가 (15,80)의 값을 갖는다면 kn 모델은 무엇으로 예측하는가?
kn.predict([[15, 80]])
```

```python
## 확인 문제
kn = KNeighborsClassifier()
kn.fit(fish_data, fish_target)

for n in range(5, 50) :
  # k-최근접 이웃 개수 설정
  kn.n_neighbors = n
  # 정확도 계산
  score = kn.score(fish_data, fish_target)
  # 100% 정확도 아래로 내려가는 순간의 이웃 개수와 정확도 출력
  if score < 1:
    print(n, score)
    break
```

```python
import pandas as pd

# CSV 불러오기
df = pd.read_csv("fish.csv")

# Smelt(빙어)는 빼고 비교하니까 제거
df = df[df['Species'] != 'Smelt']

# 입력 데이터 (Weight, Length2 사용: Length2가 평균적인 길이라 Length2 사용)
fish_data = df[['Weight', 'Length2']].to_numpy()

# 레이블: 도미(Bream)=1, 나머지=0
fish_target = (df['Species'] == 'Bream').astype(int).to_numpy()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# CSV 불러오기
df = pd.read_csv("fish.csv")

# Smelt(빙어) 제거
df = df[df['Species'] != 'Smelt']

# 비교할 대상 물고기 종
species_list = ["Roach", "Whitefish", "Parkki", "Perch", "Pike"]

plt.figure(figsize=(15, 4))

for i, sp in enumerate(species_list):
    # 도미와 현재 종만 선택
    sub = df[(df['Species'] == "Bream") | (df['Species'] == sp)]

    # 입력 데이터: Weight, Length2
    fish_data = sub[['Weight', 'Length2']].to_numpy()
    # 레이블: 도미=1, 다른 종=0
    fish_target = (sub['Species'] == "Bream").astype(int).to_numpy()

    # 학습/테스트 분할
    X_train, X_test, y_train, y_test = train_test_split(
        fish_data, fish_target, stratify=fish_target, random_state=42
    )

    # 모델 학습
    kn = KNeighborsClassifier(n_neighbors=5)
    kn.fit(X_train, y_train)

    # 정확도 출력
    print(f"{sp} vs Bream → 훈련 정확도: {kn.score(X_train, y_train):.2f}, 테스트 정확도: {kn.score(X_test, y_test):.2f}")

    # 시각화
    plt.subplot(1, len(species_list), i+1)
    plt.scatter(X_train[:,1], X_train[:,0], c=y_train, marker='o')
    plt.scatter(X_test[:,1], X_test[:,0], c=y_test, marker='^')
    plt.xlabel("Length2")
    plt.ylabel("Weight")
    plt.title(f"Bream vs {sp}")

plt.tight_layout()
plt.show()
```
